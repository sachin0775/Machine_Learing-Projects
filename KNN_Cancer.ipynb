```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
```

```python
lc = load_breast_cancer()
```

```python
df = pd.DataFrame(data=lc.data,columns=lc.feature_names)
```

```python
df['Target'] = lc.target
```

```python
df['Name_of_Cancer'] = df.Target.apply(lambda x: lc.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['Target','Name_of_Cancer'],axis=1)
x.shape
```

```python
y = df.Target
```

```python
x1 = (x- np.min(x))/ (np.max(x) - np.min(x))
x1
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y,test_size=.3,random_state=42)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(x_test)
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
sns.heatmap(data=cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
